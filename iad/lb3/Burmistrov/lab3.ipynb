Поиск лица на фотографии

```

def face_search(img):
    found_face = DeepFace.detectFace(img, detector_backend = 'mtcnn')
    plt.imshow(cv2.cvtColor(found_face, cv2.COLOR_BGR2RGB))
```

Вывод фотографии

```
def face_search_original(img):
    original = cv2.imread(img)
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
```

Сравнение фотографий

```
def compare(img1, img2):
    obj = DeepFace.verify(img1, img2, distance_metric='euclidean', model_name='DeepFace', detector_backend = 'mtcnn')
    print(obj)
```
